```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])

display(data.info())
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek
display(data.head())
```

1. You need to calculate the monthly sales of the store and identify which month had the
highest sales and which month had the lowest sales.

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
fig = px.line(sales_by_month,
x='Order Month',
y='Sales',
title='Monthly Sales Analysis')

fig.show()
```

Findings: November is the peak sales month (likely holiday/Black Friday season), while February records the lowest sales.

```python
sales_by_month
```

2. You need to analyze sales based on product categories and determine which category has
the lowest sales and which category has the highest sales.

```python
sales_by_category= data.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
fig= px.pie(sales_by_category,
            values='Sales',
            names='Category',
            hole=0.5,
            color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside',textinfo='percent+label')
fig.update_layout(title_text='Sales_Analysis_by_Category', title_font=dict(size=24))
fig.show()
```

Findings: Technology leads in sales. Office Supplies has the lowest — though it sells frequently, items are low-value.

```python
fig= px.pie (profit_by_category,
             values= 'Profit',
             names= 'Category',
             hole=0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo= 'percent+label')
fig.update_layout(title_text='Profit_Analaysis_by_Category',title_font= dict(size=24))
fig.show()
```

6. Analyze the sales and profit by customer segment

```python
sales_profit_by_segment= data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()


color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
y=sales_profit_by_segment['Sales'],
name='Sales',
marker_color=color_palette[0]))
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
y=sales_profit_by_segment['Profit'],
name='Profit',
marker_color=color_palette[1]))
fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
xaxis_title='Customer Segment', yaxis_title='Amount')

fig.show()
```

### Findings for Sales and Profit by Customer Segment:

*   **Consumer** is the leading segment in both total sales and profit.
*   **Home Office** contributes the least to both sales and profit.

```python
sales_profit_by_segment
```

7. Analyze the sales to profit ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment[ 'Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```

Findings for Sales to Profit Ratio:
Home Office shows the most efficient profit generation (ratio 7.13).
Corporate is next (ratio 7.68).
Consumer has the highest sales to profit ratio (8.66), meaning lower profitability per sale despite high overall sales.
